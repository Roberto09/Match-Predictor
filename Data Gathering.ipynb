```python
def get_data(rows, jornada, ano, tipo, res, lst):
    for (i, row) in enumerate(rows):
        cells = row.find_all("td")
        for j in range(1, 11):
            res[lst[j-1]].append(cells[j].get_text())
            if j == 1:
                res[lst[j-1]][-1] = int(res[lst[j-1]][-1])
            elif j == 2:
                res[lst[j-1]][-1] = str.lower(unidecode.unidecode(res[lst[j-1]][-1]))
            else:
                res[lst[j-1]][-1] = float(res[lst[j-1]][-1])
    for i in range(len(rows)):
        res['JRN'].append(int(jornada))
        res['ANO'].append(int(ano))
        res['TIPO'].append(tipo)
```

```python
def get_data_url(URL, jornada, ano, tipo, res):
    r = requests.get(URL)
    soup = BeautifulSoup(r.content, 'html5lib')
    table = soup.find(attrs = {'class':'sticky-enabled'})
    table = soup.find('tbody')
    rows = table.find_all("tr")
    lst = ['POS', 'NM', 'JJ', 'JG', 'JE', 'JP', 'GF', 'GC', 'DIF', 'PTS']
    get_data(rows, jornada, ano, tipo, res, lst)
```

```python
def get_data(lnk):
    with urllib.request.urlopen(lnk) as url:
        data = json.loads(url.read().decode())
        return data
```

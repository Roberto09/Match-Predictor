```python
def get_data(rows, jornada, ano, tipo, res, lst):
    for (i, row) in enumerate(rows):
        cells = row.find_all("td")
        for j in range(1, 11):
            res[lst[j-1]].append(cells[j].get_text())
            if j == 1:
                res[lst[j-1]][-1] = int(res[lst[j-1]][-1])
            elif j == 2:
                res[lst[j-1]][-1] = str.lower(unidecode.unidecode(res[lst[j-1]][-1]))
            else:
                res[lst[j-1]][-1] = float(res[lst[j-1]][-1])
    for i in range(len(rows)):
        res['JRN'].append(int(jornada))
        res['ANO'].append(int(ano))
        res['TIPO'].append(tipo)
```

```python
def get_data_url(URL, jornada, ano, tipo, res):
    r = requests.get(URL)
    soup = BeautifulSoup(r.content, 'html5lib')
    table = soup.find(attrs = {'class':'sticky-enabled'})
    table = soup.find('tbody')
    rows = table.find_all("tr")
    lst = ['POS', 'NM', 'JJ', 'JG', 'JE', 'JP', 'GF', 'GC', 'DIF', 'PTS']
    get_data(rows, jornada, ano, tipo, res, lst)
```

```python
import numpy as nm
teams = []
for team in texp.NM.unique():
    team_df = texp[texp.NM == team]
    team_df['sort_val'] = team_df.ANO + team_df.JRN/1000 + (team_df.TIPO == 'apertura')/10
    team_df = team_df.sort_values('sort_val').drop('sort_val', 1)
    teams.append(team_df)
```

```python
n = 10
for t_df in teams:
    t_df.reset_index(drop = True, inplace = True)
    clmns = ['JG', 'JE', 'JP', 'GF', 'GC', 'DIF', 'PTS']
    for clmn in clmns: t_df['R_' + clmn + '_JJ'] = -1.

    for i in range(len(t_df)):
        for clmn in clmns:
            ttl_dif = 0
            for j in range(max(0, i-n), i):
                if(j == 0):
                    dif = t_df.loc[j, clmn]
                else:
                    dif = t_df.loc[j, clmn] - t_df.loc[j-1, clmn]
                    dif = dif if dif >= 0 else t_df.loc[j, clmn]
                ttl_dif += dif
            t_df.loc[i, 'R_' + clmn + '_JJ'] = ttl_dif / min(n, i+1)
```
